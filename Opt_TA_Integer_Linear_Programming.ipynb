```python
# Creates a list of worker bool 
worker_bool = []
for worker in workers['Freelancer_username']:
    worker_bool.append(worker)
worker_bool
```

### Create Worker-Task-Values DataFrame

```python
# duplicate the tasks by the number of workers 
in_task = []
for task in tasks['taskID']:
    #for i in range(0,len(worker_bool)):
        in_task.append(task)
in_task
```

```python
# create task values DataFrame
values = pd.DataFrame(columns=['taskID','Freelancer_username'])
# populate  the dataFrame
for task in in_task:
    for worker in worker_bool:
        values = values.append({'taskID':task, 'Freelancer_username':worker}, ignore_index=True)

values
```

```python
values.set_index(['taskID','Freelancer_username'],inplace=True) 
values.index
```

#### we calculate a task value for each incoming task. 

#### The task value is associated with the estimated  accuracy, skill fitness and cost of the task. The task value is calculated as a weighted linear combination of estimated accuracy,  skill fitness (higher is better) and cost (lower is better).

#### populate  the values dataframe

```python
import numpy as np
value =  np.random.random((80,))

values['task_value'] = value
       

values
```

##  Model our decision variables using the LpVariable class

```python
worker_status = pulp.LpVariable.dicts("worker_status",
                                     ((task_ID, Freelancer_username) for task_ID, Freelancer_username in values.index),
                                     cat='Binary')
```

### instantiate a problem class, weâ€™ll name it â€œLP_Optimization_Problemâ€ and weâ€™re looking for an optimal maximum so we use LpMaximize

# Objective Function:  //////////////////////  

```python
LP_Optimization_model = pulp.LpProblem("LP Optimization Problem", pulp.LpMaximize)
```

```python
LP_Optimization_model += pulp.lpSum(
    [worker_status[task_ID, Freelancer_username] * values.loc[(task_ID, Freelancer_username), 'task_value'] 
     for task_ID, Freelancer_username in values.index])
```

#### Build constraints

```python
 #number of tasks that the worker can be assigned to  must be between 0 and 3, or zero.
for task_ID, Freelancer_username in values.index:
    min_worker_tasks = 0
    max_worker_tasks = 1
    LP_Optimization_model += worker_status[(task_ID, Freelancer_username)] >= min_worker_tasks 
    LP_Optimization_model += worker_status[(task_ID, Freelancer_username)] <= max_worker_tasks
```

#### Solve the model 

```python
LP_Optimization_model.solve()
pulp.LpStatus[LP_Optimization_model.status]
```

### print the solution 

```python
output = []
for taskID, Freelancer_username in worker_status :
    var_output = {
        'taskID': taskID,
        'Freelancer_username': Freelancer_username,
        'worker_status': worker_status[(taskID, Freelancer_username)].varValue,
    }
    output.append(var_output)
output_df = pd.DataFrame.from_records(output).sort_values(['taskID', 'Freelancer_username'])
output_df.set_index(['taskID', 'Freelancer_username'], inplace=True)
output_df
```
